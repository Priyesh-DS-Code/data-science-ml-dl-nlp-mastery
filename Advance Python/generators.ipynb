## Generators

```python
def square(n):
    for i in range(n):
        yield i**2
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a
```

```python
next(square(3))
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 3
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

## Practical Example: Reading Large Files

```python
## Practical real-world Example - Reading Large Files

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
import os

print(os.getcwd())
```

```python
file_path = 'large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
