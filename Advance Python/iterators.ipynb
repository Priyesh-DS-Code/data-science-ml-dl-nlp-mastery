## Iterators

```python
my_list = [1,2,3,4,5]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
## Iterators
iterator = iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
## Iterate through all the elements

next(iterator)
```

```python
iterator = iter(my_list)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```

```python
## String Iterator

my_string = "Priyesh"
string_iterator = iter(my_string)

print(next(string_iterator))
```
