```python
sentence = "The Eiffel Tower was built from 1887 to 1889 by French engineer Gustave Eiffel, whose company pecialized in building metal frameworks and structures."
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
tag_elements = nltk.pos_tag(words)
```

```python
# nltk.download('words')
# nltk.download('maxent_ne_chunker_tab')
# !pip install svgling

nltk.ne_chunk(tag_elements)
```
