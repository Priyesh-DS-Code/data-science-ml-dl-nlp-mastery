```python
## Classification Problem
## Comments of product is a positive review or negative review

words = ['eating', 'eats', 'eaten', 'writting', 'writes', 'programming', 'programs', 'history', 'finally', 'finalized']
```

## 1. porterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()

for word in words:
    print(word + "---->" + porter_stemmer.stem(word))
```

```python
## Disadvantage: history->histori
```

```python
porter_stemmer.stem('congratulation')
```

```python
porter_stemmer.stem('sitting')
```

## 2. RegexpStemmer class

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

## 3. Snowball Stemmer

```python
from nltk.stem import SnowballStemmer 
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word + "---->" + snowball_stemmer.stem(word))
```

```python
porter_stemmer.stem('fairly'), porter_stemmer.stem('sportingly')
```

```python
snowball_stemmer.stem('fairly'), snowball_stemmer.stem('sportingly')
```

```python
snowball_stemmer.stem('goes')
```
