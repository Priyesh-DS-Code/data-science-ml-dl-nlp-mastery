```python
corpus = '''Hello Welcome, to Priyesh's Github Repositories.
Please go through with each of my learning! to understand the NLP.
'''
```

```python
print(corpus)
```

```python
## tokenization
## paragraph --> Sentences 

import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt_tab')
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## tokenization 
## paragraph -> words
## sentence -> words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))    
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)

## now we can split priyesh's -> 'priyesh',  "'", 's'
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)

## it only takes the last fullstop as fullstop, not mid of the fullstop
```
